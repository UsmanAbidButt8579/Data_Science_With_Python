```python
# import libraries

import pandas as pd
import numpy as  np
import matplotlib.pyplot as plt
```

```python
# import online data
from sklearn.datasets import load_digits
digits=load_digits()
```

```python
# nput variables or features (X)
digits.data.shape
X=digits.data
# means 1797 are pictures and size 64=8x8
```

```python
# output labels (y)
digits.target.shape
y=digits.target
```

```python
plt.figure(figsize=(20,4))
for index, (image,label) in enumerate(zip(digits.data[0:10],digits.target[0:10])):
    plt.subplot(1,10,index +1)
    plt.imshow(np.reshape(image,(8,8)), cmap=plt.cm.gray)
    plt.title('Training: %i\n' % label, fontsize=20)
```

```python
plt.figure(figsize=(20,4))
for index, (image,label) in enumerate(zip(digits.data[0:5],digits.target[0:5])):
    plt.subplot(1,5,index +1)
    plt.imshow(np.reshape(image,(8,8)), cmap=plt.cm.gray)
    plt.title('Training: %i\n' % label, fontsize=10)
```

```python
plt.figure(figsize=(20,4))
for index, (image,label) in enumerate(zip(digits.data[0:20],digits.target[0:20])):
    plt.subplot(1,20,index +1)
    plt.imshow(np.reshape(image,(8,8)), cmap=plt.cm.gray)
    plt.title( label, fontsize=20)
```

```python
# help(plt)
```

```python
# split the data 
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y, test_size=0.2,random_state=0)
```

```python
print("Train input data: ", X_train.shape)
print("Test input data: " ,X_test.shape)
print("Train output data: ", y_train.shape)
print("Test output data: " ,y_test.shape)
```

```python
# model train
from sklearn.linear_model import LogisticRegression

model=LogisticRegression().fit(X_train,y_train)
model
```

```python
model.predict(X_test[0:10])

predictions=model.predict(X_test)
predictions
```

```python
# Accuracy test
score=model.score(X_test,y_test)
print("The accuracy score is:",score)
```

```python
# Confusion metrix
from sklearn import  metrics

cm=metrics.confusion_matrix(y_test,predictions)
cm
```

```python
import seaborn as sns
plt.figure(figsize=(9,9))
sns.heatmap(cm,annot=True,fmt=" .3f",linewidths=.5,square=True,cmap='Spectral');
plt.ylabel('Actual output');
plt.xlabel('Predicted output');
all_sample_title='Accuracy Score:{0}'.format(score)
plt.title(all_sample_title,size=15)

```

```python
# getting missed classified label
import numpy as np
import matplotlib.pyplot as plt
index=0
misclassifIedindexes=[]
for label,predict in zip(y_test,predictions):
    if label !=predict:
        misclassifIedindexes.append(index)
    index+=1
```

```python
# plotting misclassified labels with known lables
plt.figure(figsize=(20,4))
for plotIndex, badIndex in enumerate(misclassifIedindexes[0:5]):
    plt.subplot(1,5,plotIndex +1)
    plt.imshow(np.reshape(X_test[badIndex],(8,8)), cmap=plt.cm.gray)
    plt.title("Predicted: {}, Actual: {}".format(predictions[badIndex],y_test[badIndex]), fontsize=20)
```
