#3-Univariate Analysis

Analyze each feature individually to understand its distribution and characteristics:

- Use `.value_counts()` to find the frequency of unique values in a column
- Create histograms and density plots using `plt.hist()` and `sns.displot()` to visualize the distribution of numerical features [1][3]
- For categorical features, use bar plots with `plt.bar()` or count plots with `sns.countplot()` [3]

```python
# find the frequency of unique values in a column

df['sex'].value_counts()
```

```python
# Create histograms and density plots using `plt.hist()` and `sns.displot()` to visualize the distribution of numerical features

plt.hist(df['age'])
```

```python
# density plot
sns.displot(df['age'], kde=True)
```

```python
# For categorical features, use bar plots with `plt.bar()` or count plots with `sns.countplot()

sns.countplot(x='sex', data=df)
```

```python
# For categorical features, use bar plots with `plt.bar()`

sns.barplot(x='sex', y='age', data=df)
```

#4-Bivariate Analysis

Explore the relationship between two variables:

- Create scatter plots using `plt.scatter()` or `sns.scatterplot()` to visualize the relationship between two numerical features
- Use box plots with `sns.boxplot()` to compare the distribution of a numerical feature across categories
- Generate correlation matrices using `.corr()` and visualize them with `sns.heatmap()` to find linear relationships between numerical features

```python
# Create scatter plots

sns.scatterplot(x='age', y='fare', data=df)
```

```python
# Create box plots

sns.boxplot(x='sex', y='age', data=df)
```

```python
# creat correlation matrices between numerical features

df[['fare','age','sibsp','parch']].corr()
```

```python
# Plot the heatmap

sns.heatmap(df[['fare','age','sibsp','parch']].corr())
```

#5-Handling Missing Values

- Identify the percentage of missing values in each column using `.isnull().mean()` [1]
- Handle missing values by dropping rows or columns with `dropna()`, or impute them using mean, median, or other strategies

```python
df.isnull().mean()
```

```python
#  impute age and fare column null values using mean and update dataset

df['age'] = df['age'].fillna(df['age'].mean())
```

```python
df.isnull().sum()
```

```python
# drop the deck column

df.drop('deck', axis=1, inplace=True)
```

```python
df.isnull().sum()
```

```python
# drop all the null values 

df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
# show me the unique values

df['age'].unique()
```

```python
# remove unique values

df.drop_duplicates(inplace=True)
```

```python
# show me duplicates rows

df.duplicated()
```
