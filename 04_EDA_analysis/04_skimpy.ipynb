# How to do skimming of the data?

Skimming is a quick way to get a feel for the data. It is a way to get a sense of the structure, size, and types of the data. Here are some steps to do skimming of your data:

1. Import the necessary libraries.
2. Load your data into a pandas DataFrame.
3. Use the `skimpy.skim()` function to get a summary of your data.
4. Use the `skimpy.describe()` function to get a detailed summary of your data.
5. Use the `info()` function to get the data types and non-null count of each column.
6. Use the `head()` function to see the first few rows of the data.
